Merges multiple .txt files into one

```python
#selecting the data
import os

all_lines = []

for filename in os.listdir("/content/drive/MyDrive/artem-yushko/data-artem/raw/target/"):
   with open(os.path.join("/content/drive/MyDrive/artem-yushko/data-artem/raw/target/", filename), 'r') as f:
     text = f.read()
     lines = text.split('\n')
     all_lines.append(lines)

print(len(all_lines))
```

```python
# combining the raw sentences together
sentences_raw = []

for i in range(len(all_lines)):
  for j in range(len(all_lines[i])):
    if all_lines[i][j] != '' and all_lines[i][j] != '\n':
      sentences_raw.append(all_lines[i][j])

# removing the duplicates
sentences_raw = list(set(sentences_raw))
print(len(sentences_raw))
```

```python
#saving the lines to a new file
with open("/content/drive/MyDrive/artem-yushko/data-artem/raw/borshch2.txt", 'w') as f:
  for sentence in sentences_raw:
      f.write(sentence + '\n')
```
